```python
df.sort_values(['Type 1', 'HP'], ascending=[1,0]) #type 1 is going a-z (ascending = true) and HP is descending (ascending = false, ie 0)
```

```python
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#make a new column that sums all of the stats

# df = df.drop(columns=['Total'])
# delete a column, remove it from memory

df['Total'] = df.iloc[:, 4:10].sum(axis=1)
#make a new column from the sum in all rows (:,) of the columns index [4:10] in a new column (axis=1)

cols = list(df.columns)
#get the columns as a list
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
#re-arrange the order of the columns, [0:4] then the final column, then [4:12]
#note the square brackets around cols[-1]; it is a string, not a list without the additional square brackets

df.head(5)
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
#filtering for grass type, poison type and HP greater than 70
# or = |
#assigning this filter to a new df, i.e. new_df

new_df.reset_index(drop=True, inplace=True)
#resets the index values, drop the old index and change the value within the new_df as opposed to creating a new df with these parameters

new_df

new_df.to_csv('filtered.csv')
#send the new df to a csv
print(new_df)
```
